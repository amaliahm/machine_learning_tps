6. Apply dummy encoding to all categorical features in the dataset.

<div/>

```python
categorical_columns = df.select_dtypes( include = ['object'] ).columns
categorical_columns
```

```python
df_encoded = pd.get_dummies(
    df, columns = categorical_columns, drop_first = True
)

df_encoded.info()
```

```python
df_encoded.head()
```

```python
for column in df_encoded.columns:
    df_encoded[column] = df_encoded[column].astype(int)
```

8. Display the distribution of all features on all the scaled data (except binary features) using a histplot. What do you observe?

<div/>

```python
binary_features = df_encoded
binary_features
```

```python
df_encoded.info()
```

```python

# Step 2: Select numerical features excluding binary features
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns
non_binary_features = [feature for feature in numerical_features if feature not in binary_features]

# Step 3: Scale the non-binary features
scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[non_binary_features] = scaler.fit_transform(df[non_binary_features])

# Step 4: Plot the distribution of each scaled feature (excluding binary features)
plt.figure(figsize=(12, 12))
for i, feature in enumerate(non_binary_features, 1):
    plt.subplot(len(non_binary_features) // 3 + 1, 3, i)  # Adjust the grid size dynamically
    sns.histplot(df_scaled[feature], kde=True, bins=20, color='blue')
    plt.title(f'Distribution of {feature}')
    plt.tight_layout()

plt.show()
```
