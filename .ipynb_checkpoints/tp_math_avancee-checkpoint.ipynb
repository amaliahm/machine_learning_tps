```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# import
df = pd.read_csv('dataset/World_Population.csv')

# print first 10 samples
df
```

```python
df.columns
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Select relevant columns
data = df[['Time', 'TPopulation1July']]

# Drop rows with missing values
data = data.dropna()
data.head()
```

```python

# Check the data
print(data.head())

# Prepare the independent (X) and dependent (y) variables
X = data['Time'].values.reshape(-1, 1)  # Year
y = data['TPopulation1July'].values    # Population

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Evaluate the model
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"Slope (m): {model.coef_[0]}")
print(f"Intercept (b): {model.intercept_}")

# Plot the data and regression line
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Growth Prediction')
plt.legend()
plt.show()
```
